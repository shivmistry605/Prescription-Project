```python
import matplotlib.pyplot as plt
import csv
```

```python
x=[]
y=[]
```

```python
with open('azure_presc_4.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(str(row[0]))
        y.append(100*float(row[1]))
```

```python
plt.plot(x,y, marker='o')
plt.title('Prescription 4 Azure')
plt.xlabel('Words')
plt.ylabel('Confidence Score')
plt.show()
```
